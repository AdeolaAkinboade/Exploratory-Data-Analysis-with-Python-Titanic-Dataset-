# IMPORTING LIBRARIES

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# Reading the Dataset

```python
train= pd.read_csv(r'C:\Users\Jennifer\Desktop\titanic_train.csv')
```

```python
train.head()
```

# Hypothesis :
# Members that are in the cheapest Passeneger Class are less likely to survive
# More males than females are less likely to survive

# Checking for Null values(Missing Data)

```python
train.isnull()
```

```python
sns.heatmap(train.isnull(),yticklabels=False,cbar=False,cmap='viridis')
plt.title('The Categories that Contains NaN values')
```

# Checking to see survived the titanic vs who didn't survive on titanic

```python
sns.set_style('darkgrid')
sns.countplot(x='survived',data=train)
plt.title('Members OF titanic who survived vs those who did not survive')
```

# Checking more in depth to see survived vs non-survivors based on sex

```python
sns.set_style('darkgrid')
sns.countplot(x='survived',hue='sex',data=train)
plt.title('Survivor count vs non-survivor count based on Sex')
```

# Checking survivors vs non-survivors based on passenger class

```python
sns.set_style('darkgrid')
sns.countplot(x='survived',hue='pclass',data=train)
plt.title('Survivors vs Non-survivors based on Passenger Class')
```

# Using Distribution graphs to analyze age

```python
# Removing 'NaN foound in the Age column for more accuruate data'
new_train = train.dropna(subset=['age'])
```

```python
new_train
```

```python
sns.displot(data=new_train,x='age',kind='hist',aspect=1.4,bins=40)
plt.title('Distribution graph based on ages')
```

# Count Plot of Members Who Had Sibling or Spouse

```python
sns.countplot(x='sibsp',data=new_train)
plt.title('Count Plot of Members Who had Sibling or Spouse')
sns.set_style('darkgrid')
```

# Average Fare 

```python
sns.displot(data=new_train,x='fare',kind='hist',aspect=1.4,color='purple',bins=40)
plt.title('Average Fare')
```

# Relationship Between Pasenger class and Ages

```python
plt.figure(figsize=(12,8))
sns.boxplot(x='pclass',y='age',data=new_train,palette='autumn')
```

# Relationship Between Passenger Class and Fare

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='pclass',y='fare',data=new_train,palette='summer')
plt.title('Relationship Betwwn Passenger Class and Fare')
```
